```python
f = open("GM12878_OE_KRnorm_1M_chr2.txt","r")

file = f.readlines()
contact = []
col1 = []
col2 = []

size = 243
matrix = zeros((size-1,size-1))

for i in file:
    contact.append(float(i.split("\t")[2][:-2]))
    col1.append(float(i.split("\t")[0]))
    col2.append(float(i.split("\t")[1]))

# find where the strange gap is
for i in range(len(col1)-1):
    diff = col1[i+1] -col1[i]
    if diff > 1000000:
        pass
print(len(contact))

p = 0       
for j in range(size-1):
    for k in range(j+1):
        matrix[j][k] = contact[p]
        matrix[k][j] = contact[p]
        p += 1
        

    
```
